```python
from datasets import Dataset
from mutopia import parse_rdf_metadata, filter_instrument
```

```python
# Notes for Mutopia data cleaning

# Currently accepted meta_data tags:
# - "Piano"
# - "Classical Guitar"
# - "Guitar"
# - "Organ"
# - "Voice (SATB)"
# - "Harpsichord, Piano"
# - "String Quartet" - OK if "score" found in name
# - "String Quartet: Two Violins, Viola, 'Cello" - OK if "score" found in name
# - "Voice and Piano"
# - "Piano, Pianoforte, Harpsichord, Clavichord" 
# - "Voice"
# - "Piano Duet"
# - "Harpsichord"
# - "Ensemble: Mandolin, 2 Violins, 'Cello" - OK if "score" found in name 
# - "Harpsichord, Piano, Clavichord"
# - "Violin, Viola"
# - "Harpsichord,Clavichord"
# - "Violin and Piano" - OK if "score" found in file name
# - "Choir (SATB)"
# - "Clavier"
```

```python
dataset = Dataset.build(
    dir="../data/raw/mutopia",
    recur=True, 
    metadata_fn=parse_rdf_metadata,
    filter_fn=filter_instrument,
)
```

```python
md_dict = {}
for p_roll in dataset.train + dataset.test:
    instrum = p_roll.meta_data['for']
    md_dict[instrum] = md_dict.get(instrum, 0) + 1
    
for k, v in sorted(md_dict.items(), key = lambda item: -item[1]):
    print(k, v)
```

```python
for p_roll in dataset.train + dataset.test:
    if p_roll.meta_data['for'] == "Choir (SATB)":
        print(p_roll.meta_data['composer'], p_roll.meta_data['title'], p_roll.meta_data['file_name'])

```
